### Read data from all 5 files

```python
Al_df = pd.read_csv('Al_cond.csv')  # ensure path is correct
Ni_df = pd.read_csv('Ni_cond.csv')
Fe_df = pd.read_csv('Fe_cond.csv')
Cu_df = pd.read_csv('Cu_cond.csv')
W_df = pd.read_csv('W_cond.csv')
combined_df = pd.read_csv('combined_cond.csv')
```

### Visualize
____

```python
# Al
plt.figure(figsize=(10,6))
plt.scatter(Al_df["T(K)"], Al_df["log10(con)"], c=Al_df["Z"], label = "Al")
plt.xlabel("T [K]")
plt.ylabel("$\log(\sigma)$")
plt.grid(alpha = 0.3)
plt.text(27000, 3.3, "Al", fontsize=40, backgroundcolor = 'w')
```

```python
# Ni
plt.figure(figsize=(10,6))
plt.scatter(Ni_df["T(K)"], Ni_df["log10(con)"], c=Ni_df["Z"], label = "Ni")
plt.xlabel("T [K]")
plt.ylabel("$\log(\sigma)$")
plt.grid(alpha = 0.3)
plt.text(35000, 3.2, "Ni", fontsize=40, backgroundcolor = 'w')
```

```python
# Fe
plt.figure(figsize=(10,6))
plt.scatter(Fe_df["T(K)"], Fe_df["log10(con)"], c=Fe_df["Z"], label = "Fe")
plt.xlabel("T [K]")
plt.ylabel("$\log(\sigma)$")
plt.grid(alpha = 0.3)
plt.text(35000, 3, "Fe", fontsize=40, backgroundcolor = 'w')
```

```python
# Cu
plt.figure(figsize=(10,6))
plt.scatter(Cu_df["T(K)"], Cu_df["log10(con)"], c=Cu_df["Z"], label = "Cu")
plt.xlabel("T [K]")
plt.ylabel("$\log(\sigma)$")
plt.grid(alpha = 0.3)
plt.text(35000, 3, "Cu", fontsize=40, backgroundcolor = 'w')
```

```python
# W
plt.figure(figsize=(10,6))
plt.scatter(W_df["T(K)"], W_df["log10(con)"], c=W_df["Z"], label = "W")
plt.xlabel("T [K]")
plt.ylabel("$\log(\sigma)$")
plt.grid(alpha = 0.3)
plt.text(35000, 3.5, "W", fontsize=40, backgroundcolor = 'w')
```
