Connect the PandasGUI app to the electrical conductivity dataset. Requires:
1. `pandasgui`,
2. access to the CSV data files,
3. (optional) use of a virtual environment.

Info on `pandasgui` can be found [here](https://github.com/adamerose/PandasGUI).

There are many customizations you can do (e.g., change the arrangment of windows) - I recommend watching the video [here](https://pypi.org/project/pandasgui/#about). 

```python
from pandasgui import show as pg_show
import pandas as pd
import seaborn as sns
```

There isn't much to do. Run the next cell, which will bring up a GUI (look for the panda icon in your bar if nothing pops up). Nothing will be in the GUI yet, since you did not pass anything into the "show" command; but, you can use the drop downs to navigate to your file (e.g., `W.csv`) and open it. From there, it's pretty self explanatory. Probably the most useful are the plots, so click on the "Grapher" tab and start visualizing. 

You don't really need this notebook if this is all you need to do: simply start python at the command line in your directory and start `pandasgui`.

In what follows, some examples of data preparation are given to illustrate how you might prepare the data (e.g., combine datasets) before passing it directly to `pandasgui`. 

Let's read in the data, combine it, view it first through `seaborn` and then pass it to `pandasgui`. 

```python
df_W = pd.read_csv('W_cond.csv')
df_W['Z_nuc'] = 74
df_Cu = pd.read_csv('Cu_cond.csv')
df_Cu['Z_nuc'] = 29
df_Fe = pd.read_csv('Fe_cond.csv')
df_Fe['Z_nuc'] = 26
df_Ni = pd.read_csv('Ni_cond.csv')
df_Ni['Z_nuc'] = 28
df_Al = pd.read_csv('Al_cond.csv')
df_Al['Z_nuc'] = 13
```

```python
df_combined = pd.concat([df_Al, df_Fe, df_Ni, df_Cu, df_W], ignore_index=True)
```

```python
sns.pairplot(df_combined, hue="Z_nuc")
```

Now, simply call the GUI using the customized dataframe. 

```python
pg_show(df_combined)
```
